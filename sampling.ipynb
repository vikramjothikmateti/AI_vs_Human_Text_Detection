```python
import pandas as pd
```

```python
df = pd.read_csv('AI_Human.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
human = df[df['generated'] == 0.0]
ai = df[df['generated'] == 1.0]
```

```python
human_sample = human.sample(n=25000, random_state=42)
ai_sample = ai.sample(n=25000, random_state=42)
```

#### Concatenate and shuffle the new sample data from both classes

```python
AI_Human_sample = pd.concat([human_sample, ai_sample])
```

```python
AI_Human_sample.head()
```

```python
AI_Human_sample.info()
```

```python
AI_Human_sample = AI_Human_sample.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
AI_Human_sample.head()
```

```python
AI_Human_sample.info()
```

#### Save dataframe as new CSV file to continue pre processing steps

```python
AI_Human_sample.to_csv('AI_Human_50k.csv', index=False)
```
